```python
import os
from pathlib import Path

import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
project_path = "..\\Thesis_Code"
os.chdir(project_path)
```

```python
csv_file_path = "ts_times.csv"

df = pd.read_csv(csv_file_path)
ts_data = df.set_index("instance").to_dict(orient = 'index')

ts_data_final = dict((key,value) for key, value in ts_data.items())


sic_a = []
sic_b = []

fun_a = []
fun_b = []

ex_a = []
ex_b = []

ts_sic_a = []
ts_sic_b = []

for key, _ in ts_data_final.items():
    sic_a.append(ts_data_final[key]["sic_a"])
    sic_b.append(ts_data_final[key]["sic_b"])
    
    fun_a.append(ts_data_final[key]["fun_a"])
    fun_b.append(ts_data_final[key]["fun_b"])
    
    ex_a.append(ts_data_final[key]["ex_a"])
    ex_b.append(ts_data_final[key]["ex_b"])
    
    ts_sic_a.append(ts_data_final[key]["ts_sic_a"])
    ts_sic_b.append(ts_data_final[key]["ts_sic_b"])
    
```

```python
def make_graph2():
    
    x_titles = [r'$k=4$' + "\n" + '$\pi(i)=A$',
                r'$k=4$' + "\n" + '$\pi(i)=B$',
                r'$k=5$' + "\n" + '$\pi(i)=A$',
                r'$k=5$' + "\n" + '$\pi(i)=B$',
                r'$k=6$' + "\n" + '$\pi(i)=A$',
                r'$k=6$' + "\n" + '$\pi(i)=B$']
    width =0.08
    
    
    fig, ax1 = plt.subplots(figsize=(12,8))
    plt.rcParams.update({'font.size': 14})
    
    ax1.set_title("CTDS vs. TS")
    plt.xticks(range(len(sic_a)), x_titles)
    
    
    
    
    ax1.set_ylabel("cost of single solution " + r'$\mathcal{C}(\pi(i), i)$')
    ax1.set_xlabel("instances")
    
    
    b1 = ax1.bar(np.arange(len(fun_a)) - 5*width, fun_a, width=width, color="cornflowerblue",
                 label="CTDS " + r'$\mathcal{C}(\pi(i), i)$' + " with " + r'$speed_A$')
    b2 = ax1.bar(np.arange(len(fun_b)) - 4*width, fun_b, width=width, color="darkblue",
                 label="CTDS " + r'$\mathcal{C}(\pi(i), i)$' + " with " + r'$speed_B$')
    
    
    
    
    
    
    ax2=ax1.twinx()
    ax2.set_ylabel("SIC")
    
    
    
    
    b3 = ax2.bar(np.arange(len(sic_a)) - 2*width, sic_a, width=width, color="mediumseagreen",
                 label="CTDS SIC with " + r'$speed_A$')
    b4 = ax2.bar(np.arange(len(sic_b)) - 1*width, sic_b, width=width, color="darkgreen",
                 label="CTDS SIC with " + r'$speed_B$')
    
    
    b5 = ax2.bar(np.arange(len(ts_sic_a)) + 1*width, ts_sic_a, width=width, color="lightcoral",
                 label="TS SIC with " + r'$speed_A$')
    b6 = ax2.bar(np.arange(len(ts_sic_b)) + 2*width, ts_sic_b, width=width, color="darkred",
                 label="TS SIC with " + r'$speed_B$')
    
   
   
    

    
   
    
    
    ax1.set_ylim([0, 20])
    ax2.set_ylim([0, 75])
    plt.grid()
    #ax1.legend(loc="upper left")
    #ax2.legend(loc="upper center")
    
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc="center left", bbox_to_anchor=(1.08, 0.5))
    
    
    #ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    
    
    #plt.show()
    
    
    plt.savefig('ctds_vs_ts.pdf', bbox_inches='tight')

make_graph2()
```
